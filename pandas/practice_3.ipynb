```python
air_quality_pm25 = pd.read_csv("air_quality_pm25_long.csv", parse_dates=True)
air_quality_pm25 = air_quality_pm25[["date.utc","location", "parameter", "value"]]
air_quality_pm25.head()
```

```python
import pandas as pd

air_quality = pd.read_csv("air_quality_no2_long.csv")

air_quality = air_quality.rename(columns={"date.utc": "datetime"})

air_quality.head()
```

```python
import pandas as pd

air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])

air_quality["datetime"].head()
```

```python
air_quality["datetime"].min()
```

```python
air_quality["datetime"].max()
```

```python
air_quality["datetime"].max() - air_quality["datetime"].min()
```

```python
air_quality["month"] = air_quality["datetime"].dt.month
air_quality.head()
```

```python
air_quality.groupby(
    [air_quality["datetime"].dt.weekday, "location"])["value"].mean()
```

```python
import matplotlib.pyplot as plt 

fig, axs = plt.subplots(figsize=(12, 4))

air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(

    kind='bar', rot=0, ax=axs

)

plt.xlabel("Hour of the day");

plt.ylabel("$NO_2 (µg/m^3)$");
```

```python
no_2 = air_quality.pivot(index="datetime", columns='location', values="value")
no_2.head()
```

```python
no_2.index.year
```

```python
no_2.index.month
```

```python
no_2.index.weekday
```

```python
no_2["2019-05-20":"2019-05-22"].plot()
```

```python
monthly_max = no_2.resample("M").max()

monthly_max
```

```python
monthly_max.index.freq
```

```python
no_2.resample("D").mean().plot(style="-o", figsize=(10, 5));
```
